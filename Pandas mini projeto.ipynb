# Mini Projeto de Análise de Dados


### O que temos?

Temos os dados de 2019 de uma empresa de prestação de serviços. 

- CadastroFuncionarios
- CadastroClientes
- BaseServiçosPrestados


### O que queremos saber/fazer?

1. Valor Total da Folha Salarial -> Qual foi o gasto total com salários de funcionários pela empresa? <br>  

    
2. Qual foi o faturamento da empresa? <br>

    
3. Qual o % de funcionários que já fechou algum contrato?<br>

    
4. total de contratos que cada área da empresa já fechou


5.  o total de funcionários por área


6. Qual o ticket médio mensal (faturamento médio mensal) dos contratos?<br>


```python
## IMPORTANDO ARQUIVOS
import pandas as pd
funcionarios_df = pd.read_csv('CadastroFuncionarios.csv', sep=';', decimal=',')
servicos_df = pd.read_excel('BaseServiçosPrestados.xlsx')
clientes_df = pd.read_csv('CadastroClientes.csv', sep=';', decimal=',')
display(funcionarios_df, servicos_df, clientes_df)
```

### 1 - Folha salarial
- total gasto com salários + benefícios + impostos dos funcionários

```python

funcionarios_df = pd.read_csv('CadastroFuncionarios.csv', sep=';', decimal=',')    

funcionarios_df['Salario Total'] = funcionarios_df['Salario Base']  + funcionarios_df['Beneficios'] \
+ funcionarios_df['VT'] + funcionarios_df['VR'] + funcionarios_df['Impostos']

display(funcionarios_df)

print('salario total é : {:,.2f}'.format(funcionarios_df['Salario Total'].sum()))
```

### 2- Faturamento
- faturamento total de todos os serviços

```python
faturamento_df = servicos_df.merge(clientes_df, on='ID Cliente')
faturamento_df = faturamento_df.dropna(how='all', axis=1)
faturamento_df['Faturamento'] = faturamento_df['Tempo Total de Contrato (Meses)'] * faturamento_df['Valor Contrato Mensal']
faturamento_df = faturamento_df.drop(['Codigo do Servico', 'ID Funcionário'], axis=1)
display(faturamento_df)
print('o faturamento total foi de : {:,}'.format(faturamento_df['Faturamento'].sum()))
```

###  3-  % de funcionários que fecharam contrato

```python
qnt_func_fecharamcontrato = len(servicos_df['ID Funcionário'].unique())
qnt_func_total = len(funcionarios_df['ID Funcionário'])
print('Há {} funcionários na empresa'.format(qnt_func_total))
print('{} funcionários fecharam contratos'.format(qnt_func_fecharamcontrato))
print('{:.2%} dos funcionários fecharam contrato'.format(qnt_func_fecharamcontrato / qnt_func_total))
```

### 4 - total de contratos que cada área da empresa já fechou

```python
contratosarea_df = servicos_df[['ID Funcionário']].merge(funcionarios_df[['ID Funcionário', 'Area']], on='ID Funcionário')
display(contratosarea_df)
print(contratosarea_df['Area'].value_counts())
contratosarea_df['Area'].value_counts().sort_values().plot(kind= 'barh')
```

### 5 - Funcionários por área
- quantidade de funcionários por área

```python
print(funcionarios_df['Area'].value_counts())
```

### 6- faturamento médio mensal
- faturamento médio mensal dos contratos

```python
print('O faturamento médio mensal dos contratos foi de R$ {:.2f}'.format(clientes_df['Valor Contrato Mensal'].mean()))
```
